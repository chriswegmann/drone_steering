## Usage of the DataEnsembler class with timebased labels

```python
from app.module import DataEnsembler
```

```python
# initialize the DataEnsembler instance with an assumed average framelength - default is 120 ms
de = DataEnsembler(ms_per_frame = 120)
```

```python
# the DataEnsembler expects certain filenames
DataEnsembler.pattern
```

```python
# first you need to look for files that match the Regex-Pattern in a specified directory
# as a default, the DataEnsembler instance will look for timebased labels
de.investigate_available_datafiles("data/gesture/")
```

```python
# the directory name will be stored in the data_directory field
de.data_directory
```

```python
# all filenames that match the pattern will be stored to the data_source_df dataframe
de.data_source_df
```

```python
# the DataEnsembler instance tries to pair feature and label files; the results are stored to the combined_data_files dataframe
de.combined_data_files_df
```

```python
# after features and labels have been paired, the data can be loaded
de.load_data()

# this is the raw data, the labels are not vectorized and the training data is not yet in 3D-shape
# the labels will be stored to "labels"
print(de.labels[5])

# the samples will be stored to "data"
de.data[0].head()
```

```python
# Then the final data is assembled 
# The DataEnsembler uses multiple instances of the LabelGenerator class with tolerance_range and max_error as specified
#  --> See notebook "LabelGenerator_(timebased)""
de.assemble_data(tolerance_range=600,max_error=500)
```

```python
# you can display a summary of the assembled data
de.display_information()
```

```python
# the final training data is stored in "X". The shape is [sample size] x [timesteps per sample] x [number of features] 
de.X.shape
```

```python
# the final labels are encoded in "y"
print(de.y[0:100])
```
